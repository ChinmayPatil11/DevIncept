## Checking relationship between all the variables using pairplot

```python
sns.pairplot(df, vars=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)','petal width (cm)'],hue = 'target')
```

## Finding correlations using HeatMap

```python
sns.heatmap(df.corr(), annot=True)
```

We can see that the features 'petal length (cm)' and 'petal width (cm)' are highly correlated with each other which tells that they increase or decrease somewhat together. Also, sepal width is correlated to both these features to some extent. The 'sepal width (cm)' feature on the other hand is not correlated to any other feature

```python
columns_list = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)','petal width (cm)']
plt.figure(figsize=(24,4))
i = 1
for col in columns_list:
    plt.subplot(1,6,i)
    sns.distplot(df[col])
    i = i + 1
plt.tight_layout()
```

```python
iris_setosa_df = df[df['target'] == 0]
iris_versicolor_df = df[df['target'] == 1]
iris_virginica_df = df[df['target'] == 2]
```

```python
sns.distplot(iris_setosa_df['sepal length (cm)'],label = 'setosa')
sns.distplot(iris_versicolor_df['sepal length (cm)'],label = 'versicolor')
sns.distplot(iris_virginica_df['sepal length (cm)'],label = 'virginica')
plt.legend()
```

From the distribution plot of sepal length, any particular conclusion can't be inferred. Virginica species has sepal length greater than 7 cm whereas the other 2 species are less than that this length which can be inferred.

```python
sns.distplot(iris_setosa_df['sepal width (cm)'],label = 'setosa')
sns.distplot(iris_versicolor_df['sepal width (cm)'],label = 'versicolor')
sns.distplot(iris_virginica_df['sepal width (cm)'],label = 'virginica')
plt.legend()
```

From the distribution plot of sepal width, no particular conclusion can be inferred. Every species has around the same range of sepal width which is from 2 cm to 4.5 cm.

```python
sns.distplot(iris_setosa_df['petal length (cm)'],label = 'setosa')
sns.distplot(iris_versicolor_df['petal length (cm)'],label = 'versicolor')
sns.distplot(iris_virginica_df['petal length (cm)'],label = 'virginica')
plt.legend()
```

From the distribution plot of petal length it is evident that setosa species has very small petal length which totally distinguishes it from versicolor and viriginca species. Versicolor has petal length between 3 cm to around 5 cm. Virginica has petal length between 4.5 cm to 7 cm.

```python
sns.distplot(iris_setosa_df['petal width (cm)'],label = 'setosa')
sns.distplot(iris_versicolor_df['petal width (cm)'],label = 'versicolor')
sns.distplot(iris_virginica_df['petal width (cm)'],label = 'virginica')
plt.legend()
```

From the distribution plot of petal width it is evident that setosa species has very small petal width which totally distinguishes it from versicolor and viriginca species. Versicolor has width between 1 cm to around 1.7 cm and virginica has width between 1.5 cm to 2.5 cm.
